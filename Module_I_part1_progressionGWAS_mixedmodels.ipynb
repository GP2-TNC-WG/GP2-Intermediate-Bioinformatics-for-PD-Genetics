## Background/Motivation:

Genome-wide association studies (GWASs) are a method to test common genetic variants across the genome. In Parkinson's disease (PD), there have been a number of large GWASs comparing PD cases to controls to identify variants associated with disease risk. 

However, it is also important to study the genetics and biology of disease progression. We are now able to do this with increasing data available from longitudinal PD studies with detailed clinical data. By conducting progression GWASs on PD cases only, we can study the association between genetic variants and rate of progression, or other PD phenotypes. These studies can provide insight into genetic variants, genes, and pathways which might influence the rate of progression and could be targeted to develop new disease-modifying treatments.

There are different ways to measure and analyse progression, for example, depending on whether the outcome is continuous or binary (yes/no). Here we are using mixed effects models to analyse a longitudinal continuous outcome.

Run mixed effects model just on one SNP first. See https://cran.r-project.org/web/packages/lme4/vignettes/lmer.pdf for options on specifying random effect terms.

```python
%%R

#Fit mixed effects model, adjusting for sex, age at onset, and PC1-5
fit <- lmer(UPDRSIII_total ~ `1:693731_G` + SEX + age_onset + PC1 + PC2 + PC3 + PC4 + PC5
                            + `1:693731_G`*years_from_BL
                            + (years_from_BL | ID), data = merged)

summary(fit)
```

### Make function to run mixed effects models on multiple SNPs

```python
%%R

#Function to run mixed effects models analysis on each SNP
run_mixed_model = function(x){
    
    #Get the SNP name
    snp = SNPs[x]
    
    #Fit mixed effects model, adjusting for sex, age at onset, and PC1-5
    fit <- lmer(UPDRSIII_total ~ merged[[snp]] + SEX + age_onset
                            + PC1 + PC2 + PC3 + PC4 + PC5
                            + merged[[snp]]*years_from_BL
                            + (years_from_BL | ID), data = merged)

    #Get coefficients from mixed effects model
    coef_mat = summary(fit)$coef

    #Get results for the SNP (the last row)
    res = coef_mat[11,]
    
    #Number of observations (longitudinal)
    n_obs <- length(merged[[snp]])
    
    #Number of individuals
    n_inds <- sapply(ranef(fit),nrow)
    
    #Calculate minor allele frequency
    
    #First get the number of unique individuals (remove multiple rows for each individual)
    #Remove missing genotypes
    unique_ids <- merged %>%
        distinct(IID, .keep_all = T) %>% 
        filter(!is.na(SNPs[x]))
    
    #Calculate MAF - sum of the SNP counts divided by number of chromosomes (number of individuals x 2)
    maf <- sum(unique_ids[[snp]])/(nrow(unique_ids)*2)
    
    #Make sumstats final list - SNP name, number of observations (longitudinal), pvalue,
    sumstat = cbind(SNPs[x], #SNP name
                    n_obs, #number of observations (longitudinal)
                    res[1], #beta 
                    res[2], #SE
                    res[3], #DF
                    res[4], #t-value
                    res[5], #p-value
                    n_inds, #number of individuals
                    maf) #minor allele frequency
    
    return(sumstat)
}
```

### Run mixed effects models on test 20 SNPs

```python
%%R

#Get the SNP names from the genetic data - excluding the first 6 columns which are FID, IID, sex, pheno, etc.
SNPs = names(genetic)[-c(1:6)]

print(SNPs)
```

```python
%%R

#Apply function to the dataset. Make temp results dataframe
temp = lapply(1:length(SNPs), run_mixed_model)

temp
```

```python
%%R

#Make second temporary results dataframe which binds list into a dataframe
temp2 <- do.call(rbind, temp) %>%
    data.frame

#Set column names for the results dataframe
colnames(temp2) <- c("POS_A1", "OBS", "BETA", "SE", "DF", "Tvalue", "P", "N", "ALT_Frq")
#Set row names to null 
rownames(temp2) <- NULL

#Make final results dataframe including the A1 allele
results_final <- temp2 %>%
separate(POS_A1, into = c("SNP", "A1"), sep = "_")

print(results_final)

#Save final results dataframe
write.table(results_final, "/home/jupyter/progressionGWAS_results/mixedmodels/test_snps.txt",
           quote = F, row.names = F, col.names = T, sep = "\t")
```

**Structure of task**

**Scatter-gather** is where you can perform jobs in parallel (instead of sequentially). You run the same task on each input. This is a useful way to make your jobs finish faster.

More useful information about scatter/gather (i.e. running the same process on lots of inputs, e.g. in a GWAS):
* https://support.terra.bio/hc/en-us/articles/360037128572-Scatter-gather-parallelism
* https://support.terra.bio/hc/en-us/articles/360037493811--4-howto-Use-scatter-gather-to-joint-call-genotypes


Illustration of scatter-gather (from https://support.terra.bio/hc/en-us/articles/360037128572-Scatter-gather-parallelism)
